```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
```

# Hanning Window

```python
nsamp = 100 # number of sampling in $\Omega$ space...faking a continuous function
Homega = [0.]*nsamp # Frequency response
Ndow = 35
hn=[0]*Ndow # initialize window size
for n in range(-Ndow//2+1,Ndow//2+1):    
    hn[n+Ndow//2] = 0.5 + 0.5*np.cos(2*np.pi*n/(Ndow-1)) # Hanning Window function


ii = 0 ## Counter for building the frequency response
while ii < nsamp:
    i=0
    omega = ii*np.pi/nsamp ## converting counter to a digital frequency (zero to pi)
    while i < Ndow:
        Homega[ii] += hn[i]*cmath.exp(complex(0,-(i*omega))) #building H at an Omega
        i+=1
    ii+=1

plt.title('Frequency Response of a Hanning Window Filter')
plt.plot(np.abs(Homega))
plt.show()
```

# Hamming Window

```python
nsamp = 100 # number of sampling in $\Omega$ space...faking a continuous function
Homega = [0.]*nsamp # Frequency response
Ndow = 35
hn=[0]*Ndow # initialize window size
for n in range(-Ndow//2+1,Ndow//2+1):    
    hn[n+Ndow//2] = 0.54 + 0.46*np.cos(2*np.pi*n/(Ndow-1)) # Hamming Window function


ii = 0 ## Counter for building the frequency response
while ii < nsamp:
    i=0
    omega = ii*np.pi/nsamp ## converting counter to a digital frequency (zero to pi)
    while i < Ndow:
        Homega[ii] += hn[i]*cmath.exp(complex(0,-(i*omega))) #building H at an Omega
        i+=1
    ii+=1

plt.title('Frequency Response of a Hamming Window Filter')
plt.plot(np.abs(Homega))
plt.show()
```

# Blackman Window

```python
nsamp = 100 # number of sampling in $\Omega$ space...faking a continuous function
Homega = [0.]*nsamp # Frequency response
Ndow = 35
hn=[0]*Ndow # initialize window size
for n in range(-Ndow//2+1,Ndow//2+1):    
    hn[n+Ndow//2] = 0.5 + 0.5*np.cos(2*np.pi*n/(Ndow-1)) + 0.08*np.cos(4*np.pi*n/(Ndow-1)) # Blackman Window function


ii = 0 ## Counter for building the frequency response
while ii < nsamp:
    i=0
    omega = ii*np.pi/nsamp ## converting counter to a digital frequency (zero to pi)
    while i < Ndow:
        Homega[ii] += hn[i]*cmath.exp(complex(0,-(i*omega))) #building H at an Omega
        i+=1
    ii+=1

plt.title('Frequency Response of a Blackman Window Filter')
plt.plot(np.abs(Homega))
plt.show()
```
