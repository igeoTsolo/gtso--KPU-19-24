# Quantization Error

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,9)

plt.axis([0,8,0,8])
plt.xlabel("Analog Sample Value")
plt.ylabel("Quantized Level")
plt.plot(x)
plt.step(x,x-1)
plt.show()
```

```python
#Quantization Error vs Analog sample value
y_ana = np.arange(0, 9, 0.1)

quantized = y_ana//1
qerror = quantized - y_ana

plt.axis([0,8,-1,0])
plt.xlabel("Analog Sample Value")
plt.ylabel("Quantization Error")
plt.plot(y_ana, qerror)
```

Maximum error = -1

```python
x=np.arange(0,9)

plt.axis([0,8,0,8])
plt.xlabel("Analog Sample Value")
plt.ylabel("Quantized Level")
plt.plot(x+0.5, x)
plt.step(x,x-1)
plt.show()
```

```python
#Half Quantization Error vs Analog Sample Value
#Quantization Error vs Analog sample value
y_ana = np.arange(0, 9, 0.1)

quantized = y_ana//1 +0.5
qerror = quantized - y_ana

plt.axis([0,8,-1,1])
plt.xlabel("Analog Sample Value")
plt.ylabel("Quantization Error")
plt.plot(y_ana, qerror)
```

Maximum error (magnitude) = 0.5

# Sampling

```python
A=2 # Amplitude of the sin wave
f=5.0 # frequency of the sin wave in Hz
T=1/f

fana=100.0*f
Tana=1/fana # horizontal spacing of the points that "fake" a continuous signal

cycles=50 # how many cycles of the wave to save

t=np.arange(0,cycles*T,Tana) # horizontal coordinate of your sin curve
x=A*np.sin(2*np.pi*f*t) # vertical value of your sin curve

plt.axis([0,2,-2,2]) # were not plotting all cycles so we can see whats going on
plt.plot(t,x, label = 'Analog Sample')





fsamp=30.0
Tsamp=1/fsamp # horizontal spacing of the sampling points

t_samp=np.arange(0,cycles*T,Tsamp) # horizontal coordinate of your sin curve
xsamp30=A*np.sin(2*np.pi*f*t_samp) # vertical value of your sin curve
plt.axis([0,2,-2,2]) # were not plotting all cycles so we can see whats going on
plt.stem(t_samp,xsamp30, linefmt = 'r')

plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title("sin Curve")

plt.figure(figsize=(12,4))
plt.show()
```

# Aliasing

```python
fft = np.fft.fft(xsamp30) #xsamp is your stem plot data
N = xsamp30.size
Fsamp = 30
f = np.linspace(0, Fsamp, N)
#Tsamp is the rate you sampled at (30 Hz, 12 Hz, 10 Hz, 7 Hz)

plt.ylabel("Amplitude")

plt.xlabel("Frequency [Hz]")
plt.stem(f,np.abs(fft)/N)
plt.show()

```

```python
fft = np.fft.fft(xsamp7) #xsamp is your stem plot data
N = xsamp7.size
Fsamp = 7
f = np.linspace(0, Fsamp, N)
#Tsamp is the rate you sampled at (30 Hz, 12 Hz, 10 Hz, 7 Hz)

plt.ylabel("Amplitude")

plt.xlabel("Frequency [Hz]")
plt.stem(f,np.abs(fft)/N)
plt.show()

```

The far right peak moves towards the left. All 4 FFT plots show a highest peak at 5Hz, which is expected, and until the sampling frequency no longer meets the conditions of the Nyquist Sampling theorem (f_sampling must be greater than 2  5Hz = 10Hz), this is true.

When the sampling frequency is exactly 10, every sample is almost exactly on the t-axis, so the fourier transform probably failed to show significant data, hence the mess of peaks with low amplitudes. The tallest peak does still land on 5Hz, however.

When the sampling frequency is lower than 10 (7Hz), 5Hz becomes the far right peak, and the left peak becomes the "incorrect" one.
