```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ws = np.loadtxt("weatherstats_vancouver_daily.csv", delimiter=",", usecols=(17), skiprows=(1))
print(ws)
```

```python
print("Wind Speed Mean: %3.3f km/h" % (np.mean(ws)))
```

```python
plt.hist(ws, bins='auto')
plt.xlabel("Wind Speed (km/h)")
plt.ylabel("Occurences (hours)")
plt.title("Daily wind speed distribution from 1991 to 2018 in Vancouver")
plt.show()
```

```python
plt.hist(ws, bins=100)
plt.xlabel("Wind Speed (km/h)")
plt.ylabel("Occurences (hours)")
plt.title("Daily wind speed distribution from 1991 to 2018 in Vancouver")
plt.show()
```

```python
n, bins, patches=plt.hist(ws, 100)
plt.setp(patches[14], 'facecolor', 'r')
plt.text(30, 200, r'$\bar{x}=$%3.3f $\sigma=$%5.2f' % (np.mean(ws), np.std(ws)))
plt.axis([0, 50, 0, 700])
plt.grid(True)
plt.xlabel("Wind Speed (km/h)")
plt.ylabel("Occurences (hours)")
plt.title("Daily wind speed distribution from 1991 to 2018 in Vancouver")
plt.show()
```

```python
evendist = np.random.randint(0, 100, 10000)
plt.hist(evendist, bins = 50)
plt.xlabel("Random Number")
plt.ylabel("Count")
plt.title("Evenly Distributed Random Numbers from 0-100")
plt.show()
```

```python
normaldist = np.random.normal(50, 2, 10000)
plt.hist(normaldist, bins = 50)
plt.xlabel("Random Number")
plt.ylabel("Count")
plt.title("Normally Distributed Random Numbers from 0-100")
plt.show()
```

# Central Limit Theorem

```python
x1 = np.random.randint(0, 100, 10000)
x2 = np.random.randint(0, 100, 10000)

normalsum = x1+x2
plt.hist(normalsum, bins = 100)
plt.xlabel("Random Number")
plt.ylabel("Count")
plt.title("Adding 2 Random Numbers Together")
plt.show()
```

```python
x1 = np.random.randint(0, 100, 10000)
x2 = np.random.randint(0, 100, 10000)
x3 = np.random.randint(0, 100, 10000)
x4 = np.random.randint(0, 100, 10000)
x5 = np.random.randint(0, 100, 10000)
x6 = np.random.randint(0, 100, 10000)
x7 = np.random.randint(0, 100, 10000)
x8 = np.random.randint(0, 100, 10000)
x9 = np.random.randint(0, 100, 10000)
x10 = np.random.randint(0, 100, 10000)
x11 = np.random.randint(0, 100, 10000)
x12 = np.random.randint(0, 100, 10000)

normalsum = x1+x2+x3+x4+x5+x6+x7+x8+x9+x10+x11+x12

plt.hist(normalsum, bins = 100)
plt.xlabel("Random Number")
plt.ylabel("Count")
plt.title("Adding 12 Random Numbers Together")
plt.show()
```
