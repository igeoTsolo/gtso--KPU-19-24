#### 4.2 Urgg, I hate those negative values and extreme values, let's get rid of them and call it a new catalogue as 'cat_non_noex' and replot

```python
cat_non_noex = cat[(cat['MASS']<5000) & (cat['MASS']>0) &
                   (cat['L_BOL']<8000) & (cat['L_BOL']>0)]

sb.pairplot(cat_non_noex[['MASS', 'TEMP', 'L_BOL', 'T_BOL', 'SURF_DENS']])
```

#### 4.3 Use hue to distinguish

```python
sb.pairplot(cat_non_noex[['MASS', 'TEMP', 'L_BOL', 'T_BOL', 'SURF_DENS', 'EVOL_FLAG']], hue = 'EVOL_FLAG')
```

Do you see any correlation between the columns??
#### 4.4 create a correlation matrix

```python
corr = cat[['MASS', 'TEMP', 'L_BOL', 'T_BOL', 'SURF_DENS', 'EVOL_FLAG']].corr()
corr
```

#### 4.5 create a heatmap of the correlation matrix

```python
plt.figure(figsize=(10,8))
sb.heatmap(corr, annot=True, linewidth=1, linecolor='w', cmap='coolwarm_r')
```

#### 4.6 create a clustermap of the correlation matrix

```python
plt.figure(figsize=(10,8))
sb.clustermap(corr, annot=True, linewidth=1, linecolor='w', cmap='coolwarm_r')
```
