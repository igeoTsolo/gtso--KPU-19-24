#### 2.3 CHALLENGING: Use subplot to plot the histogram of F70, F160, F250, F350, F500, F870 (see the image below)
- Review you knowledge on plt.subplots
- Use what you did for 2.2 (sns.histplot)
- Use two for loops (to define i & j)
- This link might be useful https://dev.to/thalesbruno/subplotting-with-matplotlib-and-seaborn-5ei8
- **Note: we have not covered sns.histplot(ax[i,j]) and you need to use ax[i,j for this exercise]**

```python
fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(10,8))
plt.tight_layout()


a=['F70', 'F160', 'F250', 'F350', 'F500', 'F870']
a=np.reshape(a, (3,2))
b=100

for i in range(3):
    for j in range(2):
        
        sns.histplot(x=cat[a[i,j]][(cat[a[i,j]]<b)
                              & (cat[a[i,j]]>0)
                     ],
                ax=ax[i,j]
                )
```

# Act 4 - Create the following plots
#### 4.1 Create a pairplot of the following columns <br> 'MASS', 'TEMP', 'L_BOL', 'T_BOL', 'SURF_DENS'

```python
sns.pairplot(cat[['MASS', 'TEMP', 'L_BOL', 'T_BOL', 'SURF_DENS']])
```

#### 4.2 Urgg, I hate those negative values and extreme values, let's get rid of them and call it a new catalogue as 'cat_non_noex' and replot

```python
cat_non_noex = cat[(cat['MASS']>0) & (cat['MASS']<5000)
                   & (cat['L_BOL']>0) & (cat['L_BOL']<8000)
                  ]

sns.pairplot(cat_non_noex[['MASS', 'TEMP', 'L_BOL', 'T_BOL', 'SURF_DENS']])
```

#### 4.3 Use hue to distinguish

```python
sns.pairplot(cat_non_noex[['MASS', 'TEMP', 'L_BOL', 'T_BOL', 'SURF_DENS', 'EVOL_FLAG']], 
             hue='EVOL_FLAG'
            )
```

Do you see any correlation between the columns??
#### 4.4 create a corelation matrix

```python
corr = cat_non_noex[['MASS', 'TEMP', 'L_BOL', 'T_BOL', 'SURF_DENS', 'EVOL_FLAG']].corr()
corr
```

#### 4.5 create a heatmap of the correlation matrix

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, linewidth=1, linecolor='w', cmap='coolwarm_r')
```

#### 4.6 create a clustermap of the correlation matrix

```python
sns.clustermap(corr, annot=True, linewidth=1, linecolor='w', cmap='coolwarm_r')
```

What do the two plots above tell you?
