# Act 2 - Create the following plots

#### 2.1 Use object oriented method. Define 15 points from (0,10). Plot $\sqrt x$

```python
x=np.linspace(0,10,15)

fig = plt.figure()

ax = fig.add_axes([0.1,0.1,0.8,0.8])

ax.plot(x,np.sqrt(x))
```

#### 2.2. Label your axes and add title

```python
fig = plt.figure()

ax = fig.add_axes([0.1,0.1,0.8,0.8])

ax.set_xlabel('x axis')
ax.set_ylabel('$\sqrt{x}$')
ax.set_title('Root Squared Plot')

ax.plot(x,np.sqrt(x))
```

#### 2.3. Add a smaller second plot of $x^2$ on the right side of this and have appropriate labels in.

```python
fig = plt.figure()

#first plot

ax1 = fig.add_axes([0.1,0.1,0.8,0.8])

ax1.set_xlabel('x axis')
ax1.set_ylabel('$\sqrt{x}$')
ax1.set_title('Root Squared Plot')

ax1.plot(x,np.sqrt(x))

#second plot

ax2 = fig.add_axes([0.55,0.22,0.3,0.4])

ax2.set_xlabel('x axis')
ax2.set_ylabel('$x^2$')
ax2.set_title('Squared Plot')

ax2.plot(x,x**2, c='r', ls='--')
```

#### 2.4. Can you update it to look like this?

```python
fig = plt.figure()

#first plot

ax1 = fig.add_axes([0.1,0.1,0.8,0.8])

ax1.set_xlabel('x axis')
ax1.set_ylabel('$\sqrt{x}$')
ax1.set_title('Root Squared Plot')

ax1.plot(x,np.sqrt(x),
         c='c',
         lw=4,
         ls=':', 
         marker = 'X',
         markersize=10,
         markerfacecolor = 'y',
         markeredgewidth=2,
         markeredgecolor='b'
        )

#second plot

ax2 = fig.add_axes([0.55,0.22,0.3,0.4])

ax2.set_xlabel('x axis')
ax2.set_ylabel('$x^2$')
ax2.set_title('Squared Plot')

ax2.plot(x,x**2, c='r', ls='--')
```

# Act 3 - Subplots

#### Create the following plots

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=[10,4])

# This is for both plots. Note this is iteratable!
for i in range(len(ax)):
    ax[i].set_xlabel('This title is the same as the one next to it')

# This is for the first plot
ax[0].set_ylabel('$X^2$')
ax[0].plot(x,x**2,
           c='r',
           lw='8',
           alpha=0.5,
          )


# This is for the second plot
ax[1].set_ylabel('$\sqrt{X}$')
ax[1].plot(x,x**0.5)
```

#### 3.2. Add $20\cdot\sqrt{x}$ to the plot on the left, Make sure to add the legends in

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=[10,4])

# This is for both plots. Note this is iteratable!
for i in range(len(ax)):
    ax[i].set_xlabel('This title is the same as the one next to it')

# This is for the first plot
ax[0].set_ylabel('$X^2$ and $20 \cdot \sqrt{x}$')
ax[0].plot(x,x**2,
           c='r',
           lw='8',
           alpha=0.5,
           label='$X^2$'
          )

ax[0].plot(x, 20*x**0.5,
           c='k',
           lw=8,
           ls='-',
           alpha=0.5,
           label='$20 \cdot \sqrt{x}$'
          )

ax[0].legend()


# This is for the second plot
ax[1].set_ylabel('$\sqrt{X}$')
ax[1].plot(x,x**0.5)
```
