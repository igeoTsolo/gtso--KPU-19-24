# Act 1 - Read Data
#### Import numpy and pandas below

```python
import numpy as np
import pandas as pd
```

#### 1.1 Read the corrected format data "Data/HiGAL_clump_catalogue_correct_formatting.csv" and call it "cat", use read_csv

```python
cat = pd.read_csv('Data/HiGAL_clump_catalogue_correct_formatting.csv', index_col=False)
```

#### 1.2 print out the head of cat

```python
cat.head()
```

# Act 2 - Create the following plots

#### 2.1 create a histogram of the 'TEMP' (Temperature)

```python
cat['TEMP'].plot.hist()
```

Maybe 20 bins is more appropriate. Change it!

```python
cat['TEMP'].plot.hist(bins=20)
```

Create a KDE plot of this

```python
cat['TEMP'].plot.kde()
```

#### 2.2 create a histogram of the 'T_BOL' (Bolometric Temperature)

```python
cat['T_BOL'].plot.hist()
```

Maybe we don't need values bigger than 60 K. Plot for those values where smaller than 60 K

```python
cat['T_BOL'][cat['T_BOL']<60].plot.hist()
```

#### 2.3 create a histogram of the 'SURF_DENS' (Surface Density)

```python
cat['SURF_DENS'].plot.hist()
```

Plot for those values where smaller than 2 and change number of bins to 20

```python
cat['SURF_DENS'][cat['SURF_DENS']<2].plot.hist(bins=20)
```

#### 2.4 create a histogram of the 'L_BOL' (Luminosity)

```python
cat['L_BOL'].plot.hist()
```

We must have a few with the luminosity bigger than 10,000. Plot histogram where luminosity is smaller than 10,000

```python
cat['L_BOL'][cat['L_BOL']<10000].plot.hist()
```

We still have a few bigger 2000. Plot positive values where smaller than 2000!

```python
cat['L_BOL'][(cat['L_BOL']<2000) & (cat['L_BOL']>0)].plot.hist()
```

#### 2.5 create a histogram of the 'MASS' (Mass)

```python
cat['MASS'].plot.hist()
```

Plot those **positive** values smaller than 3000

```python
cat['MASS'][(cat['MASS']<3000) & (cat['MASS']>0)].plot.hist()
```

# Act 3 - Create a box plot
#### Create the following

```python
cat[['TEMP','T_BOL']].plot.box()
```

# Act 4 - Correlation between mass and luminosity

Create a scatter plot, weher x is mass, and y is luminosity

```python
cat.plot.scatter(x='MASS', y='L_BOL')
```

It does not seem to have any correlation. What if we change the colors with temperature?

```python
cat.plot.scatter(x='MASS', y='L_BOL', c='TEMP')
```

Use cmap to distinguish it better

```python
cat.plot.scatter(x='MASS', y='L_BOL', c='TEMP', cmap='coolwarm')
```
