```python
from numpy.random import rand, randn
```

```python
np.random.seed(101)
df = pd.DataFrame(np.array(randn(400)+2).reshape(100,4), columns=['150um', '250um', '350um', '500um'])
df['long'] = [10 if n == 0 else 30 for n in rand(100).round()]
df['Evolution'] = rand(100).round()
df
```

```python
df['150um'].plot.hist(bins=10)
```

```python
df['150um'].plot.kde()
```

```python
df[['150um', '250um']].plot.line()
```

```python
df.drop('long',axis = 1).plot.box()
```

```python
print(df['150um'][df['150um']<df['150um'].median()].median())
print(df['150um'][df['150um']>df['150um'].median()].median())
```

```python
df.plot.scatter(x='150um', y='250um'
               , c=df['Evolution'] # colour depends on 'evolution'
               , cmap='viridis' # colourmap
               , s = df['long']*2) # size
```

```python
import seaborn as sb
```

```python
df = pd.read_csv('C:/Users/georg/Dropbox/Kwantlen/PHYS 4900 - Introduction to Data Science/Data/CERN/dielectron.csv')
```

```python
df.head()
```

```python
df = df[(df['E1']<200) &
        (df['E2']<200) &
        (abs(df['px1 '])<50) &
        (abs(df['px2'])<50) &
        (abs(df['py1'])<50) &
        (abs(df['pz1'])<50) &
        (abs(df['py2'])<50) &
        (abs(df['pz2'])<50)
       ]

df = df[:200]
df
```

```python
sb.histplot(df['px1 '], bins = 15, kde = True)
```

```python
sb.jointplot(x = 'pt1', y = 'E1', data = df, hue = 'Q1', palette = 'coolwarm', s = df['M'])
```

```python
sb.jointplot(x = 'pt1', y = 'E1', data = df[:200], hue = 'pt1', palette = 'rainbow')
# hue set to x-axis by default
```

```python
df1 = df[['E1', 'pt1', 'eta1', 'phi1', 'Q1']]
sb.pairplot(df1)
```

```python
sb.barplot(x = 'Q1', y = 'pt1', data = df)
```
