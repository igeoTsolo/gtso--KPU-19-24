# Act 3 - Add a new column, groupby, unique and nunique
Those sources with the 8 < GLON < 12 are most likely associated with  the W31 cloud
 and those with the 28 < GLON < 32 are most likely associated with the W43.
 
#### 3.1 add a column called 'West' to your catalogue to indicate either W31 or W43 based on their GLON

```python
cat['West'] = 0

def westify(long):
    if long > 8 and long < 12:
        return('W31')
    else:
        return('W43')
    
cat['West'] = cat['GLON'].apply(westify)
```

```python
cat[['GLON','West']]
```

#### 3.2 What is the average of Surface Density in each of the W31, & W43? Hint: use groupby 

```python
cat.groupby('West').mean()['SURF_DENS']
```

#### 3.3 What is the maximum Surface Density if each of the W31, & W43?

```python
cat.groupby('West').max()['SURF_DENS']
```

#### 3.7 How many of the sources in W31 with each of the evolutionary flags?
