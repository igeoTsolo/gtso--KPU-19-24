### 1.1 - Define a Function
def a function to flip a coin! It should return "HEAD' or 'TAIL'

Hint: use randint

```python
from random import randint
def coinflip():
    coinface = randint(0,1)
    if coinface == 0:
        return 'HEADS'
    else:
        return 'TAILS'
```

```python
coinflip()
```

### 1.2 - Now flip the coin for 10 times and save the result in a list. Print the list!

```python
results = []
i=0
for i in range(10):
    coinface = coinflip()
    results.append(coinface)
    
print(results)
```

### 2.1 - Wien's Law
Look up the Wien's law! Now define a  function that accepts the input of $\lambda_{max}$ in the units of $n m$ from the peak of black body radiation and returns the temperature of the star!

```python
def WiensLaw(wavelength):
    '''wavelength is in meters'''
    propoConst = 2.898e-3 #Wien's Law constant or proportionality
    StarTemp = propoConst/wavelength*1e9
    return StarTemp
```

Call this function to find the surface temperature of the Sun, knowing that $\lambda_{Max} = 550 n m$

```python
WiensLaw(550)
```

## This following code uses a function very similar to the one you have defined above to plot the planck function for temperature of 10K!

Note: In the function wavlength must be a list!

```python
#Leave as it is

# Defining the planck function here
def planck_wavelength(wavelength, temp=2.7):
    from math import exp
    c=299792458
    h=6.62607015e-34
    k_b= 1.380649e-23
    B = list(2*h*c**2/(WAVE**5)/(exp(h*c/(WAVE*k_b*temp))-1) for WAVE in wavelength)
    return B


# Importing the required packages
import numpy as np
import matplotlib.pyplot as plt

# creating a list of values from 1e-5 to 1e-3. This range is linearly divided into 100 points 
x=np.linspace(1e-5,1e-3,100)
# Finding planck values for the x defined in the line above, at the temperature equal to 10K.
y=planck_wavelength(x,10)


## Fianally plot the function!
plt.plot(x,y)
```
