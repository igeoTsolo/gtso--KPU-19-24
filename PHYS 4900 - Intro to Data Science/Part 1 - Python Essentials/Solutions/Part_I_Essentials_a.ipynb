### Kids Magic Trick! Practice +=, -=, *=
**Ask the user to provide the 'year_of_birth'.  
Practice +=, -=, *= with the followings. Perform each task in an individual line**
- calculate user's age (check out the note below)
- Double it --> print it out and tell the user this
- add 5 --> print it out and tell the user this
- multiply by 50 --> print it out and tell the user this
- Add user's age to it --> print it out and tell the user this
- Add 365 --> print it out and tell the user this
- Subtract 615 --> print it out and tell the user this
- print out the result and tell the user the first 4 number is their year of birth and the last two is their age!

Note: to calclate the age you need to know the current year! We will learn about this later but the following cell will tell you the "current_year"


```python
## LEAVE THIS AS IT IS
from datetime import date
current_year = date.today().year
```

```python
year_of_birth = int(input("what year were you born? "))
age = current_year - year_of_birth
year_of_birth *= 2
print(f"your age x2 is: {year_of_birth}")
year_of_birth +=5
print(f"your age x2+5 is: {year_of_birth}")
year_of_birth *=50
print(f"(your age x2+5)x50 is: {year_of_birth}")
year_of_birth = year_of_birth+age
print(f"(your age x2+5)x50 + your_age is: {year_of_birth}")
year_of_birth +=365
print(f"[(your age x2+5)x50 + your_age]+365 is: {year_of_birth}")
year_of_birth -=615
print(f"[(your age x2+5)x50 + your_age]+365-615 is: {year_of_birth}")
print(f"HAHA! The first four digit is the year you were born and the last two is your age: {year_of_birth}")
```
