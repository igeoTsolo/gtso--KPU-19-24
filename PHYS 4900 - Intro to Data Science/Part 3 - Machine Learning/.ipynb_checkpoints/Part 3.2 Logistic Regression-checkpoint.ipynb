#### 2.4 Call the list above 'cols'

#### 2.5 From 'cols' we only need ['F70', 'F160', 'F250', 'F350', 'F500', 'F21', 'F22', 'F24', 'F870', 'F1100']. Delete the rest!  


```python
colsF = [item for item in cols if item[0] == 'F']
colsF
```

#### 2.6 From your data delete all the columns with EVOL_FLAG == 0
**Use this dataset for your train_test_split in the next section**

```python
catmodel = cat[cat['EVOL_FLAG'] != 0]
catmodel
```

# Act 3 - Use sklearn and Logistic Regression to train your data!
- Use cols for your Design Matrix (X)
- Your y column should be the 'EVOL_FLAG'
- Check your Coefficients and Intercept --> What do they represent?
- perform the prediction
- Create a scatter plot of $y_{test}$ vs prediction, is this representitive?
- Create confusion matrix and classification report

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x = catmodel[colsF]
x
```

```python
y = catmodel['EVOL_FLAG']
y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.33, random_state = 42)
```

```python
lg = LogisticRegression(max_iter = 10000)
```

```python
lg.fit(xtrain, ytrain)
```

```python
lg.coef_
```

```python
pd.DataFrame(lg.coef_.reshape(10,1), index = x.columns, columns = ['Coefficients']).sort_values('Coefficients', ascending = False)
```

```python
lg.intercept_
```

```python
pred = lg.predict(xtest)
```

```python
sb.scatterplot(x = pred, y = ytest)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(ytest, pred)
```

```python
print(classification_report(ytest, pred))
```

#### 4.5 Let's plot your data. Challenging!
- Each array contain 64 pixel values of a handwritten number (i.e. dg.data)
- You need to reshape them to be able to plot them!
- Review your knowledge of subplots
- You would need to use ax.imshow https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html

Can you create the following figure/plot?

```python
len(np.unique(dg.target))
```

```python
rows, cols = 10, 10

fig, ax = plt.subplots(nrows = rows, ncols = cols, figsize = (30,30))

for i in range(cols):
    for j in range(rows):
        ax[i,j].imshow(dg.data[i*10+j].reshape(8,8),  cmap='gray')
        ax[i,j].set_title(f'Target is: {dg.target[i*10+j]}')
        ax[i,j].axis('off')
```

# Act 5 -  Perform training on this dataset, and see how well the result is!

```python
x=dg.data
y=dg.target
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.33) 
```

```python
lg = LogisticRegression(max_iter = 10000)
```

```python
lg.fit(xtrain, ytrain)
```

```python
pred = lg.predict(xtest)
```

```python
print(classification_report(ytest, pred))
```

# $\LARGE Not \ Bad, \ eh? $

# Act 6 - Could you update the plot above so that for those wrong predictions we get a different cmap? Note to change the title accordingly!

```python
rows, cols = 10, 10

fig, ax = plt.subplots(nrows = rows, ncols = cols, figsize = (30,30))

for i in range(cols):
    for j in range(rows):
        if ytest[i*10+j] != pred[i*10+j]:
            ax[i,j].imshow(xtest[i*10+j].reshape(8,8),  cmap='coolwarm')
            ax[i,j].set_title(f'Target: {ytest[i*10+j]}, pred: {pred[i*10+j]}')
        else:
            ax[i,j].imshow(xtest[i*10+j].reshape(8,8),  cmap='gray')
            ax[i,j].set_title(f'Target: {ytest[i*10+j]}, pred: {pred[i*10+j]}')
        ax[i,j].axis('off')
```
