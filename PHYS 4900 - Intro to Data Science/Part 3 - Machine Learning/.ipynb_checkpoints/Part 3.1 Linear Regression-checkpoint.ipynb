# Act 3 - Linear Regression Model  & Prediction
Let's say in real life we cannot directly measure the 'stator_tooth' temperature. However, we have access to the temperature of 'coolant', 'stator_winding', & 'stator_yoke'. Can you fit a model to these to predict the temperature of 'strator_tooth'?

#### 3.1 Fit the model
- Create X & Y
- Split your data into training and set
- Import your linear regression model
- Fit your model to your data
- Check the coefficients 
- **What does each coefficient tell you?**
- What is the intercept?

```python
x = el[['coolant', 'stator_winding', 'stator_yoke']]
y = el['stator_tooth']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.33)
```

```python
lm = LinearRegression()
```

```python
lm.fit(xtrain, ytrain)
```

Each coefficient is related to which of your parameters?

```python
lm.coef_
```

```python
lm.intercept_
```

#### 3.2 Predict y$_{test}$ and evaluate your pmodel!
- Make the prediction
- A scatter plot might help to visualize your Fit. Make sure to label your axis
- Use RMSE to evaluate your fit!
- Is this a good fit? Why? Check your mean, min, & max!

```python
pred = lm.predict(xtest)
```

```python
sb.scatterplot(x = pred, y = ytest)
plt.xlabel('Predictions')
plt.ylabel('y$_{test}$')
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(ytest, pred))
```

```python
y.min(), y.max(), y.mean()
```

# Act 4 - Use all the columns except the 'profile_id' to create a linear model!
#### 4.1 Create a model and fit x-train!
- Create X & Y
- Split your data into training and set
- Fit your model to your data
- Check the coefficients 
- Make sure to create a data frame of it! It just looks nicer!
- **What does each coefficient tell you? Which one is more effective? Sort them!**
- What is the intercept?

```python
el.columns
```

```python
x = el[['u_q', 'coolant', 'stator_winding', 'u_d', 'motor_speed', 'i_d', 'i_q', 'pm', 'stator_yoke', 'ambient', 'torque']]

y = el['stator_tooth']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.33, random_state = 42)
```

```python
lm = LinearRegression()
```

```python
lm.fit(xtrain, ytrain)
```

```python
coeff = lm.coef_
coeff
```

```python
coeffdf = pd.DataFrame(data = coeff,index = [x.columns], columns = ['Coefficients'])
coeffdf
```

```python
coeffdf.sort_values(by=['Coefficients'], ascending = False)
```

```python
lm.intercept_
```

#### 4.2 Predict y$_{test}$ and evaluate your model!
- Make the prediction
- A scatter plot might help to visualize your Fit. Make sure to label your axis
- Use RMSE to evaluate your fit!
- Is this a good fit? Why? Check your mean, min, & max!
- How do you compare your RMSE to the previous Act (i.e. 3)? Does your complex model helped to improve dramatically?

```python
pred = lm.predict(xtest)
```

```python
sb.scatterplot(x = pred, y = ytest)
plt.xlabel('Predictions')
plt.ylabel('y$_{test}$')
```

```python
np.sqrt(mean_squared_error(ytest, pred))
```

```python
y.min(), y.max(), y.mean()
```
